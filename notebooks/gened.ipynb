```python
import pandas as pd

import pymongo
from dotenv import load_dotenv
import os
from pymongo import UpdateMany

load_dotenv()

mongodb_uri = os.environ.get("MONGODB_URI")

client = pymongo.MongoClient(mongodb_uri)
db = client[os.environ.get("MONGODB_DB")]
classes = db['classes']
```

```python
df = pd.read_csv('./gened-courses.csv')  
df.head()
```

```python
for index, row in df.iterrows():
    print(row['ACP'], type(row['ACP']))
```

```python
results = []
#Loop over each row in the dataframe
for index, row in df.iterrows():

    # Convert the NaN values to None
    if pd.isnull(row['ACP']):
        row['ACP'] = None
    if pd.isnull(row['CS']):
        row['CS'] = None
    if pd.isnull(row['HUM']):
        row['HUM'] = None
    if pd.isnull(row['NAT']):
        row['NAT'] = None
    if pd.isnull(row['QR']):
        row['QR'] = None
    if pd.isnull(row['SBS']):
        row['SBS'] = None
        
    gened = {
        'ACP' : row['ACP'],
        'CS' : row['CS'],
        'HUM' : row['HUM'],
        'NAT' : row['NAT'],
        'QR' : row['QR'],
        'SBS' : row['SBS'],
    }
    
    # Add the dictionary to the results list
    results.append(UpdateMany({'number' : row['Course'].split(" ")[1], 'code' : row['Course'].split(" ")[0]}, {'$set': {'gened' : gened}}))
    
```

```python
print(results[0])
```

```python
db_res = classes.bulk_write(results)
```
